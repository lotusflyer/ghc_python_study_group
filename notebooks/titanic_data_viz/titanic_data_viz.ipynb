```python
# importing required libraries
import numpy as np
import pandas as pd

# importing matplotlib
import matplotlib.pyplot as plt

# display plots in the notebook itself
%matplotlib inline
```

```python
titanic_df = pd.read_csv('titanic_data.csv')
```

```python
titanic_df
```

```python
# survivors
suvivor_gb = titanic_df.groupby('Survived').Survived.count()

# sort by sales
suvivor_gb.sort_values(inplace=True)
print(suvivor_gb)
```

```python
x = suvivor_gb.index.tolist()
y = suvivor_gb.values.tolist()

# set axis labels
plt.xlabel('Suvivor')
plt.ylabel('Count')

# set title
plt.title('Count of Survivors')

# set xticks 
plt.xticks(labels=['Survivor', 'Died'], ticks=np.arange(len(x)))

plt.bar(x, y, color=['red', 'green'])
```
